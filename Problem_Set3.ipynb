```python
def cal_capacity(z):
  per = round(z*100)
  #print(f"per: {per}")
  if per<=1:
    return("E")
  elif per>=99:
    return("F")
  else:
    return(per)

def main():
  while True:
    try:
      fraction = input("Fraction: ")
      values = fraction.split("/")
      if len(values)==2:
        x = int(values[0])
        y = int(values[1])
      else:
        continue
      if x>y or x<0 or y<=0:
        continue
      #else:
      z = x/y
      fuel = cal_capacity(z)
      if fuel=="E" or fuel=="F":
        print(fuel)
        break
      else:
        print(f"{fuel}%")
        break
    except (ValueError,ZeroDivisionError):
      pass

main()
```

```python

def main():
  months = {"January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12}
  while True:
    try:
      user_date = input("Date: ").strip()
      if "/" in user_date:
        new_date = user_date.split("/")
        if len(new_date) != 3:
          continue
        mm = int(new_date[0])
        dd = int(new_date[1])
        yyyy = int(new_date[2])
        if (1 <= mm <= 12) and (1 <= dd <= 31):
          break
      elif "," in user_date:
        month_day, year = user_date.split(",")
        year = year.strip()
        month_name, day = month_day.split()
        if (month_name in months):
          mm = months[month_name]
          dd = int(day)
          yyyy = int(year)
          if (1 <= dd <= 31):
            break
    except:
      continue
  print(f"{yyyy:04}-{mm:02}-{dd:02}")

main()
```
