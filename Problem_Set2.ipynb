```python
def main():
  amount_due = 50
  print(f"Amount Due: {amount_due}")
  while True:
    insert_coin = int(input("Insert coin: "))
    while insert_coin not in [5,10,25]:
      print(f"Amount Due: {amount_due}")
      insert_coin = int(input("Insert coin: "))
    if insert_coin >= amount_due:
      change_owed = insert_coin - amount_due
      print(f"Change Owed: {change_owed}")
      break
    elif insert_coin < amount_due:
      amount_due = amount_due - insert_coin
      print(f"Amount Due: {amount_due}")

main()

```

```python
def remove_vowels(user_string):
  new_list=[]
  for char in user_string:
    if char.lower() not in ['a','e','i','o','u']:
      new_list.append(char)
  new_string = "".join(new_list)
  return new_string

def main():
  user_string = input("Input: ")
  omitted_string = remove_vowels(user_string)
  print(f"Output: {omitted_string}")

main()


```

```python

def is_valid(s):
  number_started = False
  if not ((len(s)>=2 and len(s)<=6) and (s.isalnum())):
    return False
  if not (s[0:2].isalpha()):
    return False
  for char in s:
    if char.isdigit():
      if char == '0':
        if not number_started:
          return False
      number_started = True

    else:
      if number_started:
        return False
  return True

def main():
  plate = input("Plate: ")
  if is_valid(plate):
    print("valid")
  else:
    print("invalid")

main()
```

```python
def calculate_calories(fruit):
  fruit = fruit.lower()
  fruitcal = {"apple": 130,
              "avocado": 50,
              "banana" : 110,
              "grape" : 60,
              "orange" : 80,
              "peach" : 60,
              "pineapple" : 50

  }
  if fruit in fruitcal:
    #return fruitcal[fruit]
    return fruitcal.get(fruit)

def main():
  fruit = input("Item: ")
  calories = calculate_calories(fruit)
  if calories:
    print(f"Calories: {calories}")

main()
```
