```python
def main():
  original_string = input("Please enter your string: ")
  converted_string = convert(original_string)
  print(converted_string)

def convert(original_string):
  new_smile_string = original_string.replace(":)","😐")
  new_sad_string = new_smile_string.replace(":(","🙁")
  return new_sad_string

main()
```

```python
def calculate_energy(m):
  c = 300000000
  E = m*c**2
  return E

def main():
  mass = int(input("Please enter the value of mass: "))
  Energy = calculate_energy(mass)
  print(Energy)

main()

```

```python
def dollars_to_float(d):
  return float(d.replace("$",""))

def percent_to_float(p):
  return float(p.replace("%",""))/100

def main():
  dollars = dollars_to_float(input("How much was the meal? "))
  percent = percent_to_float(input("What percentage would you lik to tip? "))
  tip = dollars*percent
  print(f"Leave ${tip:.2f}")

main()
```
