```python
file = input("Please enter the filename: ").lower().strip()
if (file.endswith((".jpg",".jpeg"))):
  media_type = "image/jpeg"
elif (file.endswith(".gif")):
  media_type = "image/gif"
elif (file.endswith(".png")):
  media_type = "image/png"
elif (file.endswith(".pdf")):
  media_type = "application/pdf"
elif (file.endswith(".txt")):
  media_type = "text/plain"
elif (file.endswith(".zip")):
  media_type = "application/zip"
else:
  media_type = "application/octet-stream"

print(media_type)
```

```python
def get_expression():
  expression = input("Please enter the expression in the format 2 + 2: ")
  x,y,z = expression.split(" ")
  return(x,y,z)

def calculate_expression(x,y,z):
  x = float(x)
  z = float(z)
  if (y == "+"):
    result = x + z
  elif (y == "-"):
    result = x - z
  elif (y == "*"):
    result = x * z
  elif (y == "/"):
    result = x / z
  else:
    result = "Invalid expression"

  return result


def main():
  x,y,z = get_expression()
  result = calculate_expression(x,y,z)
  if (result == "Invalid expression"):
    print(result)
  else:
    print(f"{result:.1f}")

main()
```

```python
def convert_time(time):
  hour, minutes = time.split(":")
  hour = float(hour)
  minutes = float(minutes)
  minutes = (minutes/60)
  converted_time = hour + minutes
  return converted_time

def meal_type(converted_time):
  if (converted_time >= 7.00 and converted_time <= 8.00):
    meal_time = "breakfast time"
  elif (converted_time >= 12.00 and converted_time <= 13.00):
    meal_time = "lunch time"
  elif (converted_time >= 18.00 and converted_time <= 19.00):
    meal_time = "dinner time"
  else:
    return None
  return meal_time

def main():
  time = input("What time is it? ")
  converted_time = convert_time(time)
  meal_time = meal_type(converted_time)
  if meal_time:
    print(meal_time)

main()
```
